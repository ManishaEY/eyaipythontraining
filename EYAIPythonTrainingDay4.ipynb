```python
# Extract all words from the tickets
words_pattern = r"(\b[A-Za-z]+(?=[ :]))"
words = ti['ticket'].str.extractall(words_pattern)
print("Words:\n", words)
```

```python
# Extract all the USD prices
prices_pattern = r"(\$\d+(\.\d{2})?)"
prices = ti['ticket'].str.extractall(prices_pattern)
print("USD Prices:\n", prices)
```

```python
# Extract all the high-price tickets (considered as triple-digit prices)
high_prices_pattern = r"(\$\d{3,}(\.\d{2})?)"
high_prices = ti['ticket'].str.extractall(high_prices_pattern)
print("High-Price Tickets:\n", high_prices)
```
